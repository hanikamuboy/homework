[문제1]

주어진 정수 x와 자연수 n을 이용해, x부터 시작해 x씩 증가하는 숫자를 n개 지니는 리스트를 출력해주세요   

x = 2, n =5

```python
def generate_list(x, n):
    # 리스트를 생성하는 리스트 컴프리헨션
    result = [x * (i + 1) for i in range(n)]
    return result

# 예시 값
x = 2
n = 5

# 함수 호출 및 결과 출력
print(generate_list(x, n))
```

[문제2]

전화번호가 문자열 phone_number로 주어졌을 때, 전화번호의 뒷 4자리를 *으로 가린 문자열이 출력되도록 코드를 작성해주세요

phone_number = "01012347890"  

출력 예시 : "0101234****"

```python
def mask_phone_number(phone_number):
    # 전화번호의 길이 확인
    if len(phone_number) < 4:
        raise ValueError("전화번호는 4자리 이상이어야 합니다.")
    
    # 전화번호의 마지막 4자리를 '*'로 대체
    masked_part = '*' * 4
    unmasked_part = phone_number[:-4]
    
    # 결과 문자열 생성
    result = unmasked_part + masked_part
    return result

# 예시 값
phone_number = "01012347890"

# 함수 호출 및 결과 출력
print(mask_phone_number(phone_number))
```

[문제3]

숫자를 건넬 때 일부 자릿수를 영단어로 바꾼 카드를 건네주면 프로도는 원래 숫자를 찾는 게임입니다.

다음은 숫자의 일부 자릿수를 영단어로 바꾸는 예시입니다.

1478 → "one4seveneight"  
234567 → "23four5six7"  
10203 → "1zerotwozero3"    

이렇게 숫자의 일부 자릿수가 영단어로 바뀌어졌거나,   
혹은 바뀌지 않고 그대로인 문자열 s가 주어졌을때
s가 의미하는 원래 숫자를 출력하도록 코드를 작성해주세요

s = "77three4one

---
참고로 각 숫자에 대응되는 영단어는 다음 표와 같습니다.

숫자	영단어  
0	zero  
1	one  
2	two  
3	three  
4	four  
5	five  
6	six  
7	seven  
8	eight  
9	nine  

```python
def convert_string_to_number(s):
    # 영단어와 대응되는 숫자를 정의합니다.
    word_to_digit = {
        "zero": "0",
        "one": "1",
        "two": "2",
        "three": "3",
        "four": "4",
        "five": "5",
        "six": "6",
        "seven": "7",
        "eight": "8",
        "nine": "9"
    }
    
    # 영단어의 길이 리스트를 정렬합니다. (길이가 긴 단어를 먼저 처리하기 위함)
    words = sorted(word_to_digit.keys(), key=len, reverse=True)
    
    # 원래 숫자를 저장할 변수
    original_number = ""
    
    # 현재 분석 중인 문자열
    temp = s
    
    while temp:
        matched = False
        for word in words:
            if temp.startswith(word):
                original_number += word_to_digit[word]
                temp = temp[len(word):]
                matched = True
                break
        if not matched:
            # 영단어와 매칭되지 않는 부분은 그대로 추가
            original_number += temp[0]
            temp = temp[1:]
    
    return original_number

# 예시 값
s = "77three4one"

# 함수 호출 및 결과 출력
print(convert_string_to_number(s))
```

[문제4]

행렬의 덧셈은 행과 열의 크기가 같은 두 행렬의 같은 행, 같은 열의 값을 서로 더한 결과가 됩니다. 2개의 행렬 arr1과 arr2가 주어졌을 때, 행렬 덧셈의 결과를 출력하는 코드를 작성해주세요(result와 같은 결과가 출력되도록 작성)

arr1 = [[1,2],[2,3]]

arr2 = [[3,4],[5,6]]

result = [[4,6],[7,9]]

```python
def add_matrices(arr1, arr2):
    # 행렬의 행과 열의 크기 확인 (두 행렬이 같은 크기여야 함)
    if len(arr1) != len(arr2) or any(len(row1) != len(row2) for row1, row2 in zip(arr1, arr2)):
        raise ValueError("두 행렬의 크기가 같아야 합니다.")
    
    # 결과 행렬 생성
    result = [
        [arr1[i][j] + arr2[i][j] for j in range(len(arr1[0]))]
        for i in range(len(arr1))
    ]
    
    return result

# 예시 행렬
arr1 = [[1,2],[2,3]]
arr2 = [[3,4],[5,6]]

# 행렬 덧셈 수행 및 결과 출력
result = add_matrices(arr1, arr2)
print(result)
```

[문제5]

양의 정수 x가 하샤드 수이려면 x의 자릿수의 합으로 x가 나누어져야 합니다.   
예를 들어 18의 자릿수 합은 1+8=9이고, 18은 9로 나누어 떨어지므로 18은 하샤드 수입니다.   
주어진 x를 이용해 x가 하샤드 수인지 아닌지 검사하는 코드를 작성해주세요

x = 17

----

입출력 예시
x	결과값  
10	true  
12	true  
11	false  
13	false  

```python
def is_harshad_number(x):
    # x가 하샤드 수인지 여부를 검사합니다.
    
    # x의 자릿수 합을 계산합니다.
    digit_sum = sum(int(digit) for digit in str(x))
    
    # x가 자릿수 합으로 나누어 떨어지는지 검사합니다.
    return x % digit_sum == 0

# 예시 값
x = 17

# 하샤드 수 여부 검사 및 결과 출력
result = is_harshad_number(x)
print(result)
```

[문제7]

0부터 9까지의 숫자 중 일부가 들어있는 정수 배열 numbers가 주어집니다.   
numbers에서 찾을 수 없는 0부터 9까지의 숫자를 모두 찾아 더한 수를 출력해주세요

numbers = [5,8,1,0,6,9]

```python
def find_missing_sum(numbers):
    # 0부터 9까지의 모든 숫자를 포함하는 집합
    all_numbers = set(range(10))
    
    # 주어진 배열의 숫자들을 집합으로 변환
    numbers_set = set(numbers)
    
    # 포함되지 않은 숫자들을 찾기 위해 차집합을 사용
    missing_numbers = all_numbers - numbers_set
    
    # 포함되지 않은 숫자들의 합을 계산
    missing_sum = sum(missing_numbers)
    
    return missing_sum

# 예시 값
numbers = [5, 8, 1, 0, 6, 9]

# 함수 호출 및 결과 출력
result = find_missing_sum(numbers)
print(result)
```

[문제 1: 붕어빵 재고 확인]

[문제 2: 붕어빵 주문 받기]

손님이 주문한 붕어빵의 종류와 개수를 입력받는 기능을 구현하세요.

요구사항:
- 손님이 주문할 붕어빵의 종류와 개수를 입력받습니다.
- 주문 내용을 딕셔너리에 저장하세요.
- 주문 내역을 출력해주세요.

```python
# 붕어빵 재고를 딕셔너리로 관리합니다.
inventory = {
    '팥붕어빵': 10,
    '슈크림붕어빵': 8,
    '초코붕어빵': 5
}

# 주문 내역을 저장할 딕셔너리
order = {}

# 손님이 주문할 붕어빵의 종류와 개수를 입력받는 함수
def take_order(inventory):
    print("주문할 붕어빵의 종류와 개수를 입력하세요.")
    while True:
        kind = input("붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ")
        if kind == '종료':
            break
        if kind not in inventory:
            print("잘못된 붕어빵 종류입니다. 다시 입력하세요.")
            continue
        try:
            quantity = int(input(f"{kind}의 개수를 입력하세요: "))
            if quantity <= 0:
                print("개수는 1 이상의 정수여야 합니다. 다시 입력하세요.")
                continue
            if quantity > inventory[kind]:
                print(f"현재 재고가 부족합니다. {kind}의 남은 재고는 {inventory[kind]}개입니다.")
                continue
            # 주문 내역에 추가
            if kind in order:
                order[kind] += quantity
            else:
                order[kind] = quantity
            # 재고에서 주문한 개수만큼 차감
            inventory[kind] -= quantity
        except ValueError:
            print("유효한 개수를 입력하세요.")

# 주문 내역을 출력하는 함수
def display_order(order):
    if not order:
        print("주문 내역이 없습니다.")
        return
    print("주문 내역:")
    for kind, quantity in order.items():
        print(f"{kind}: {quantity}개")

# 주문을 받고 주문 내역을 출력합니다.
take_order(inventory)
display_order(order)
```

[문제 3: 붕어빵 주문 처리]

손님의 주문을 처리하여 붕어빵 재고를 업데이트하는 기능을 구현하세요.

요구사항:
- 손님의 주문 내용을 기반으로 재고를 업데이트합니다.
- 재고가 부족할 경우, 손님에게 재고 부족을 알리고 재고를 감소시키지 않습니다.
- 재고가 충분할 경우, 주문한 만큼 재고를 감소시키고 판매를 완료합니다.
- 판매가 완료된 경우 판매된 붕어빵 맛과 개수를 출력하세요.

```python
# 붕어빵 재고를 딕셔너리로 관리합니다.
inventory = {
    '팥붕어빵': 10,
    '슈크림붕어빵': 8,
    '초코붕어빵': 5
}

# 주문 내역을 저장할 딕셔너리
order = {}

# 손님이 주문할 붕어빵의 종류와 개수를 입력받는 함수
def take_order():
    print("주문할 붕어빵의 종류와 개수를 입력하세요.")
    while True:
        kind = input("붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ")
        if kind == '종료':
            break
        if kind not in inventory:
            print("잘못된 붕어빵 종류입니다. 다시 입력하세요.")
            continue
        try:
            quantity = int(input(f"{kind}의 개수를 입력하세요: "))
            if quantity <= 0:
                print("개수는 1 이상의 정수여야 합니다. 다시 입력하세요.")
                continue
            # 주문 내역에 추가
            if kind in order:
                order[kind] += quantity
            else:
                order[kind] = quantity
        except ValueError:
            print("유효한 개수를 입력하세요.")

# 재고를 업데이트하는 함수
def process_order(inventory, order):
    print("\n주문 처리 중...")
    for kind, quantity in order.items():
        if kind not in inventory:
            print(f"{kind}은(는) 재고 목록에 없습니다.")
            continue
        if quantity > inventory[kind]:
            print(f"재고 부족: {kind}의 현재 재고는 {inventory[kind]}개입니다.")
            return
        # 재고에서 주문한 개수만큼 차감
        inventory[kind] -= quantity
        print(f"{kind} {quantity}개가 판매되었습니다.")
    
    # 주문 처리 완료 후, 주문 내역을 초기화
    order.clear()

# 주문 내역을 출력하는 함수
def display_order(order):
    if not order:
        print("주문 내역이 없습니다.")
        return
    print("주문 내역:")
    for kind, quantity in order.items():
        print(f"{kind}: {quantity}개")

# 주문을 받고 주문 내역을 처리합니다.
take_order()
display_order(order)
process_order(inventory, order)
```

[문제 4: 여러 손님 주문 처리]

- 각 주문마다 재고를 업데이트합니다.
- 현재 재고를 화면에 출력합니다.

```python
# 붕어빵 재고를 딕셔너리로 관리합니다.
inventory = {
    '팥붕어빵': 10,
    '슈크림붕어빵': 8,
    '초코붕어빵': 5
}

# 손님의 주문 내역을 저장할 딕셔너리
order = {}

# 손님이 주문할 붕어빵의 종류와 개수를 입력받는 함수
def take_order():
    print("\n주문할 붕어빵의 종류와 개수를 입력하세요.")
    while True:
        kind = input("붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ")
        if kind == '종료':
            break
        if kind not in inventory:
            print("잘못된 붕어빵 종류입니다. 다시 입력하세요.")
            continue
        try:
            quantity = int(input(f"{kind}의 개수를 입력하세요: "))
            if quantity <= 0:
                print("개수는 1 이상의 정수여야 합니다. 다시 입력하세요.")
                continue
            # 주문 내역에 추가
            if kind in order:
                order[kind] += quantity
            else:
                order[kind] = quantity
        except ValueError:
            print("유효한 개수를 입력하세요.")

# 재고를 업데이트하는 함수
def process_order(inventory, order):
    print("\n주문 처리 중...")
    for kind, quantity in order.items():
        if kind not in inventory:
            print(f"{kind}은(는) 재고 목록에 없습니다.")
            continue
        if quantity > inventory[kind]:
            print(f"재고 부족: {kind}의 현재 재고는 {inventory[kind]}개입니다.")
            # 주문을 재시도하도록 하는 기능은 생략하고, 재고 부족 메시지만 표시합니다.
            return
        # 재고에서 주문한 개수만큼 차감
        inventory[kind] -= quantity
        print(f"{kind} {quantity}개가 판매되었습니다.")
    
    # 주문 처리 완료 후, 주문 내역을 초기화
    order.clear()

# 현재 재고를 출력하는 함수
def display_inventory(inventory):
    print("\n현재 붕어빵 재고:")
    for kind, quantity in inventory.items():
        print(f"{kind}: {quantity}개")

# 시스템을 반복적으로 실행하여 여러 손님의 주문을 처리합니다.
def main():
    while True:
        take_order()
        if not order:
            print("주문 내역이 없습니다.")
            continue
        process_order(inventory, order)
        display_inventory(inventory)
        # 모든 손님의 주문을 계속 받을지 여부를 결정
        more_orders = input("\n다른 손님의 주문을 계속 받으시겠습니까? (예/아니오): ")
        if more_orders.lower() != '예':
            print("시스템을 종료합니다.")
            break

# 프로그램 실행
main()
```

관리자 모드를 통해 붕어빵 재고를 추가할 수 있는 기능을 구현하세요.

요구사항:
- 관리자는 특정 붕어빵의 재고를 추가할 수 있습니다.
- "종료"를 입력하면 관리자 모드를 종료합니다.
- 추가된 재고를 업데이트하고 현재 재고를 화면에 출력합니다.

```python
# 붕어빵 재고를 딕셔너리로 관리합니다.
inventory = {
    '팥붕어빵': 10,
    '슈크림붕어빵': 8,
    '초코붕어빵': 5
}

# 손님의 주문 내역을 저장할 딕셔너리
order = {}

# 손님이 주문할 붕어빵의 종류와 개수를 입력받는 함수
def take_order():
    print("\n주문할 붕어빵의 종류와 개수를 입력하세요.")
    while True:
        kind = input("붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ")
        if kind == '종료':
            break
        if kind not in inventory:
            print("잘못된 붕어빵 종류입니다. 다시 입력하세요.")
            continue
        try:
            quantity = int(input(f"{kind}의 개수를 입력하세요: "))
            if quantity <= 0:
                print("개수는 1 이상의 정수여야 합니다. 다시 입력하세요.")
                continue
            # 주문 내역에 추가
            if kind in order:
                order[kind] += quantity
            else:
                order[kind] = quantity
        except ValueError:
            print("유효한 개수를 입력하세요.")

# 재고를 업데이트하는 함수
def process_order(inventory, order):
    print("\n주문 처리 중...")
    for kind, quantity in order.items():
        if kind not in inventory:
            print(f"{kind}은(는) 재고 목록에 없습니다.")
            continue
        if quantity > inventory[kind]:
            print(f"재고 부족: {kind}의 현재 재고는 {inventory[kind]}개입니다.")
            # 주문을 재시도하도록 하는 기능은 생략하고, 재고 부족 메시지만 표시합니다.
            return
        # 재고에서 주문한 개수만큼 차감
        inventory[kind] -= quantity
        print(f"{kind} {quantity}개가 판매되었습니다.")
    
    # 주문 처리 완료 후, 주문 내역을 초기화
    order.clear()

# 현재 재고를 출력하는 함수
def display_inventory(inventory):
    print("\n현재 붕어빵 재고:")
    for kind, quantity in inventory.items():
        print(f"{kind}: {quantity}개")

# 관리자가 재고를 추가할 수 있는 관리자 모드 함수
def admin_mode(inventory):
    while True:
        print("\n관리자 모드:")
        print("1. 재고 추가")
        print("2. 종료")
        choice = input("선택하세요 (1 또는 2): ")
        if choice == '1':
            kind = input("재고를 추가할 붕어빵 종류를 입력하세요: ")
            if kind not in inventory:
                print("잘못된 붕어빵 종류입니다.")
                continue
            try:
                additional_quantity = int(input(f"{kind}에 추가할 개수를 입력하세요: "))
                if additional_quantity <= 0:
                    print("개수는 1 이상의 정수여야 합니다. 다시 입력하세요.")
                    continue
                inventory[kind] += additional_quantity
                print(f"{kind}의 재고가 {additional_quantity}개 추가되었습니다.")
                display_inventory(inventory)
            except ValueError:
                print("유효한 개수를 입력하세요.")
        elif choice == '2':
            print("관리자 모드를 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 입력하세요.")

# 시스템을 반복적으로 실행하여 여러 손님의 주문을 처리합니다.
def main():
    while True:
        print("\n메인 메뉴:")
        print("1. 주문 받기")
        print("2. 관리자 모드")
        print("3. 종료")
        choice = input("선택하세요 (1, 2 또는 3): ")
        if choice == '1':
            take_order()
            if not order:
                print("주문 내역이 없습니다.")
                continue
            process_order(inventory, order)
            display_inventory(inventory)
        elif choice == '2':
            admin_mode(inventory)
        elif choice == '3':
            print("시스템을 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 입력하세요.")

# 프로그램 실행
main()
```

[문제 6: 매출 확인]

붕어빵 판매 가격을 설정하고, 판매한 붕어빵의 매출을 계산하세요.

요구사항:
- 각 붕어빵의 가격을 딕셔너리로 관리합니다.
- 판매된 붕어빵의 종류와 개수를 딕셔너리로 관리합니다.
- 총 매출을 계산하여 출력합니다.

초기 설정:
- 가격
    - "팥붕어빵": 1000
    - "슈크림붕어빵": 1200
    - "초코붕어빵": 1500

- 판매 개수
    - "팥붕어빵": 0
    - "슈크림붕어빵": 0
    - "초코붕어빵": 0

```python
# 붕어빵 재고를 딕셔너리로 관리합니다.
inventory = {
    '팥붕어빵': 10,
    '슈크림붕어빵': 8,
    '초코붕어빵': 5
}

# 붕어빵 가격을 딕셔너리로 관리합니다.
prices = {
    '팥붕어빵': 1000,
    '슈크림붕어빵': 1200,
    '초코붕어빵': 1500
}

# 판매된 붕어빵의 개수를 관리할 딕셔너리
sales_count = {
    '팥붕어빵': 0,
    '슈크림붕어빵': 0,
    '초코붕어빵': 0
}

# 손님의 주문 내역을 저장할 딕셔너리
order = {}

# 손님이 주문할 붕어빵의 종류와 개수를 입력받는 함수
def take_order():
    print("\n주문할 붕어빵의 종류와 개수를 입력하세요.")
    while True:
        kind = input("붕어빵 종류를 입력하세요 (종료하려면 '종료' 입력): ")
        if kind == '종료':
            break
        if kind not in inventory:
            print("잘못된 붕어빵 종류입니다. 다시 입력하세요.")
            continue
        try:
            quantity = int(input(f"{kind}의 개수를 입력하세요: "))
            if quantity <= 0:
                print("개수는 1 이상의 정수여야 합니다. 다시 입력하세요.")
                continue
            # 주문 내역에 추가
            if kind in order:
                order[kind] += quantity
            else:
                order[kind] = quantity
        except ValueError:
            print("유효한 개수를 입력하세요.")

# 재고를 업데이트하는 함수
def process_order(inventory, order, sales_count):
    print("\n주문 처리 중...")
    for kind, quantity in order.items():
        if kind not in inventory:
            print(f"{kind}은(는) 재고 목록에 없습니다.")
            continue
        if quantity > inventory[kind]:
            print(f"재고 부족: {kind}의 현재 재고는 {inventory[kind]}개입니다.")
            # 주문을 재시도하도록 하는 기능은 생략하고, 재고 부족 메시지만 표시합니다.
            return
        # 재고에서 주문한 개수만큼 차감
        inventory[kind] -= quantity
        sales_count[kind] += quantity
        print(f"{kind} {quantity}개가 판매되었습니다.")
    
    # 주문 처리 완료 후, 주문 내역을 초기화
    order.clear()

# 매출을 계산하는 함수
def calculate_revenue(sales_count, prices):
    total_revenue = 0
    for kind, quantity in sales_count.items():
        total_revenue += quantity * prices[kind]
    return total_revenue

# 현재 재고를 출력하는 함수
def display_inventory(inventory):
    print("\n현재 붕어빵 재고:")
    for kind, quantity in inventory.items():
        print(f"{kind}: {quantity}개")

# 관리자 모드를 통해 재고를 추가할 수 있는 함수
def admin_mode(inventory):
    while True:
        print("\n관리자 모드:")
        print("1. 재고 추가")
        print("2. 종료")
        choice = input("선택하세요 (1 또는 2): ")
        if choice == '1':
            kind = input("재고를 추가할 붕어빵 종류를 입력하세요: ")
            if kind not in inventory:
                print("잘못된 붕어빵 종류입니다.")
                continue
            try:
                additional_quantity = int(input(f"{kind}에 추가할 개수를 입력하세요: "))
                if additional_quantity <= 0:
                    print("개수는 1 이상의 정수여야 합니다. 다시 입력하세요.")
                    continue
                inventory[kind] += additional_quantity
                print(f"{kind}의 재고가 {additional_quantity}개 추가되었습니다.")
                display_inventory(inventory)
            except ValueError:
                print("유효한 개수를 입력하세요.")
        elif choice == '2':
            print("관리자 모드를 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 입력하세요.")

# 시스템을 반복적으로 실행하여 여러 손님의 주문을 처리합니다.
def main():
    while True:
        print("\n메인 메뉴:")
        print("1. 주문 받기")
        print("2. 관리자 모드")
        print("3. 매출 확인")
        print("4. 종료")
        choice = input("선택하세요 (1, 2, 3 또는 4): ")
        if choice == '1':
            take_order()
            if not order:
                print("주문 내역이 없습니다.")
                continue
            process_order(inventory, order, sales_count)
            display_inventory(inventory)
        elif choice == '2':
            admin_mode(inventory)
        elif choice == '3':
            total_revenue = calculate_revenue(sales_count, prices)
            print(f"\n총 매출: {total_revenue}원")
        elif choice == '4':
            print("시스템을 종료합니다.")
            break
        else:
            print("잘못된 선택입니다. 다시 입력하세요.")

# 프로그램 실행
main()
```
